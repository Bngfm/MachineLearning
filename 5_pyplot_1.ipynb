# **Pyplot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
y = x**2
x, y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y, c='red')
plt.show()
plt.close()
```

### Figure

```python
plt.figure(figsize=(12, 5))
plt.plot(x,y, c='red', label="courbe1")
plt.plot(x, x**3/5, c='black', label="courbe2")
plt.xlabel("temps")
plt.ylabel("salaire")
plt.legend()
plt.title("parametre courbe")
plt.savefig("imgJupyter.png")

plt.figure()
plt.plot(x, y)
```

## subplot


```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title("subplot(2,1,1)")

plt.subplot(2,1,2)
plt.plot(x, x*2)
plt.title("subplot(2,1,2)")

```
