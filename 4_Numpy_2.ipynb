# **Numpy manipulation**

```python
import numpy as np
```

## **Indexing and slicing**

```python
np.random.seed(0)
t = np.random.randn(4,4)
t
```

```python
t[0, 1]
```

#### First column

```python
t[:,0]
```

#### First line

```python
t[0, :]
```

#### Line precision

```python
t[1]
```

```python
t[0]
```

#### extract array

```python
t1 = t[:2,:2]
t1
```

```python
t.shape
```

```python
t1.shape
```

```python
t[:2,:2] = 0
t
```

```python
t2 = t[:,2:]
t2
```

```python
t3 = t[:,-2:]
t3
```

```python
t4 = np.zeros((5,5))
t4
```

```python
t4[::2, ::2] = 1
t4
```

# Boolean indexing

```python
A = np.random.randint(0, 100, [5,5])
A
```

```python
A[A < 50] = 100
A
```

```python
A.shape
```

```python
A[A != 100]
```

```python
A[A != 100].shape
```

# **Image exemple**

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show
```
